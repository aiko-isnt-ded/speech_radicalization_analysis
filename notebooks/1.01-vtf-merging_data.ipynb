```python
import pandas as pd
```

The data used for this project was downloaded using _Zeeschuimer_ [1], a tool developed by Digital Methods Initiative that facilitates webscrapping social media. A series of searches were performed based on common hashtags for each topic. Only the user_@ and body of the tweets was kept. Since data was webscrapped directly from Twitter, a lot of noise is to be expected in the form of emojis, URLs, mentions and more. This notebook aims to integrate and clean data before performing classification.

## <font color= #f0565e> **ICE** </font>

```python
# Load data
ice1 = pd.read_csv('../data/raw/ice/twitter_DeportThemAll.csv') 
ice2 = pd.read_csv('../data/raw/ice/twitter_MassDeportationsNow.csv')
ice3 = pd.read_csv('../data/raw/ice/twitter_ICE.csv')

df_ice = pd.concat([ice1, ice2, ice3])
df_ice = df_ice.query("datalegacylang == 'en'")
df_ice
```

## <font color= #f0565e> **LGTBQIA+ phobia** </font>

```python
# Load Data
lgbt1 = pd.read_csv('../data/raw/lgbt/twitter_GenderIdeology.csv') 
lgbt2 = pd.read_csv('../data/raw/lgbt/twitter_ParentsRights.csv') 
lgbt3 = pd.read_csv('../data/raw/lgbt/twitter_WokeMindVirus.csv') 
lgbt4 = pd.read_csv('../data/raw/lgbt/twitter_NoLGBT.csv') 

df_lgbt = pd.concat([lgbt1, lgbt2, lgbt3, lgbt4])
df_lgbt = df_lgbt.query("datalegacylang == 'en'")
df_lgbt
```

## <font color= #f0565e> **Islamophobia** </font>

```python
# Load Data
islamophobia1 = pd.read_csv('../data/raw/islamophobia/twitter_BanIslam.csv') 
islamophobia2 = pd.read_csv('../data/raw/islamophobia/twitter_IslamIsTheProblem.csv') 
islamophobia3 = pd.read_csv('../data/raw/islamophobia/twitter_StopIslam.csv') 
islamophobia4 = pd.read_csv('../data/raw/islamophobia/twitter_StopIslamInvasion.csv') 

df_islamophobia = pd.concat([islamophobia1, islamophobia2, islamophobia3, islamophobia4])
df_islamophobia = df_islamophobia.query("datalegacylang == 'en'")
df_islamophobia
```
