In order to train a custom model for our task, we first need to give it our data and label it according to the type of radical speech that it contains. These categories are:

- **Hate Speech**
- **Dangerous Speech**
- **Extreme Speech**
- **Enraged Speech**
- **Hate Grammar**
- **Horror Grammar**

Our initial dataset has over 6200 examples. Manually labeling each of these data points, considering our current time contraints, would be unfeasable. Therefore, we decided to use a Large Language Model to automate the process. We used GPT 5.4, by OpenAI. It was given the following prompt:

`I have CSVs containing the contents of specific types of radical speech in tweets. I also have these definitions for different types of radical speech. Please give a category for each speech in each of the CSV, and create new CSVs with a new column specifying their type.`

The model produced CSV files containing the classification, individually in the 5 different topics.
